```python
import pandas as pd
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load the spaCy English language model
nlp = spacy.load('en_core_web_sm')

# Load the spaCy TextBlob extension
nlp.add_pipe('spacytextblob')

# Function to preprocess text and perform sentiment analysis
def preprocess_and_analyze_sentiment(text):
    # Tokenize the text
    tokens = nlp(text.lower())
    STOP_WORDS = nlp.Defaults.stop_words
    # Remove punctuation, special characters, and stopwords
    clean_tokens = [token.text for token in tokens if token.text.isalpha() and token.text not in STOP_WORDS]
    
    # Join the tokens back into a clean text
    clean_text = ' '.join(clean_tokens)
    
    # Perform sentiment analysis using spaCy's TextBlob extension
    doc = nlp(clean_text)

    sentiment_score = doc._.polarity
    
    # Determine sentiment label based on score
    if sentiment_score > 0:
        if sentiment_score > 0.5:
            sentiment_label = 'very positive'
        else:
            sentiment_label = 'positive'
    elif sentiment_score < 0:
        if sentiment_score < -0.5:
            sentiment_label = 'very negative'
        else:
            sentiment_label = 'negative'
    else:
            sentiment_label = 'neutral'
    return sentiment_score, sentiment_label
```

Here's how to interpret VIF:

- VIF = 1: No multicollinearity. The variance of the regression coefficient is not inflated.
- VIF > 1 and < 5: Moderate multicollinearity. There may be some correlation among predictor variables, but it may not significantly affect the estimation.
- VIF > 5: High multicollinearity. There is a substantial correlation among predictor variables, which can significantly affect the estimation.

```python
num_columns = train_df.columns
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# VIF [Sequentially] Check
vif_data = train_df.copy()
total_columns = vif_data.shape[1]
column_to_be_kept = []
column_index = 0

for i in range(total_columns):
    vif_value = variance_inflation_factor(vif_data,column_index)
    print(column_index,'---------',vif_value)
    
    if vif_value <= 6:
        column_to_be_kept.append(num_columns[i])
        column_index += 1
    else:
        vif_data.drop(columns=num_columns[i],inplace=True)
```

# **Model Selection**

```python
from sklearn.model_selection import train_test_split
x = train_df.drop(columns=target_var)
# x = vif_data.drop(columns=target_var)
y = train_df[target_var]
# y = vif_data[target_var]
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb

# Initialize regression models including ensemble models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'AdaBoost': AdaBoostRegressor(),
    'XGBoost': xgb.XGBRegressor()
}

# Initialize lists to store results
model_names = []
mae_scores = []
mse_scores = []
rmse_scores = []
r2_scores = []

# Train each model, predict on test data, and evaluate performance
for name, model in models.items():
    # Train the model
    model.fit(x_train, y_train)
    
    # Predict on test data
    y_pred = model.predict(x_test)
    
    # Calculate regression metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    
    # Append results to lists
    model_names.append(name)
    mae_scores.append(mae)
    mse_scores.append(mse)
    rmse_scores.append(rmse)
    r2_scores.append(r2)

# Create DataFrame to compare model performance
results_df = pd.DataFrame({
    'Model': model_names,
    'MAE': mae_scores,
    'MSE': mse_scores,
    'RMSE': rmse_scores,
    'R-squared': r2_scores
})

# Display the results DataFrame
print(results_df)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb

# Initialize regression models including ensemble models
models = {
    'Random Forest': RandomForestRegressor(n_estimators=500),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=500),
    'XGBoost': xgb.XGBRegressor(n_estimators=500)
}

# Initialize lists to store results
model_names = []
mae_scores = []
mse_scores = []
rmse_scores = []
r2_scores = []

# Train each model, predict on test data, and evaluate performance
for name, model in models.items():
    # Train the model
    model.fit(x_train, y_train)
    
    # Predict on test data
    y_pred = model.predict(x_test)
    
    # Calculate regression metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    
    # Append results to lists
    model_names.append(name)
    mae_scores.append(mae)
    mse_scores.append(mse)
    rmse_scores.append(rmse)
    r2_scores.append(r2)

# Create DataFrame to compare model performance
results_df = pd.DataFrame({
    'Model': model_names,
    'MAE': mae_scores,
    'MSE': mse_scores,
    'RMSE': rmse_scores,
    'R-squared': r2_scores
})

# Display the results DataFrame
print(results_df)
```
